### french с PCA

```python
class HierarchicalClustering:
    
    #Instantiate the object and computes the distance matrix between the points 
    def __init__(self,dataset,method_param='ward',metric_param='euclidean'):

        #faster implementation than scipy 
        from fastcluster import  linkage_vector
        from scipy.spatial.distance import pdist

        self.Z = linkage_vector(dataset,method=method_param,metric=metric_param)

    #plot the dendrogram of the hierarchical clustering
    def dendrogram(self,max_group=10):
        from scipy.cluster.hierarchy import dendrogram
        import matplotlib.pyplot as plt
        #affichage du dendrogramme
        fig,ax=plt.subplots(figsize=(10,10))
        ax.set_title("Dendrogram")
        dendrogram(self.Z,orientation='top',truncate_mode='lastp', p=max_group,show_contracted=True)
        plt.show()


    #  Outputs the group of each point k: cluster_numbers
    def groups(self,k):
        from scipy.cluster.hierarchy import fcluster
        #cluster hierarchy by using loadings of ACP
        groupes_cah = fcluster(self.Z,k,criterion='maxclust')
        return groupes_cah
```

```python
import numpy as np
import pandas as pd
pd.set_option('display.max_columns', 300)
from scipy import stats
from scipy.cluster import hierarchy
from sklearn import ensemble, decomposition, model_selection, feature_selection, metrics, discriminant_analysis
# from lightgbm import LGBMClassifier
import sys
sys.setrecursionlimit(10000) # проблемы с иерархической кластеризацией
import matplotlib.pyplot as plt
import matplotlib.cm as cmap
```

```python
fig, axes = plt.subplots(ncols=2, nrows=2, figsize=(16, 10))
pca = decomposition.PCA(n_components=2, random_state=42)
decomposited = pca.fit_transform(X1.values)

axes[0][0].scatter(decomposited[np.where(clusters==1)[0], 0], decomposited[np.where(clusters==1)[0], 1], 
                   c='r', label='1', alpha=0.3)
axes[0][0].scatter(decomposited[np.where(clusters==2)[0], 0], decomposited[np.where(clusters==2)[0], 1], 
                   c='y', label='2', alpha=0.3)
axes[0][0].scatter(decomposited[np.where(clusters==3)[0], 0], decomposited[np.where(clusters==3)[0], 1], 
                   c='g', label='3', alpha=0.3)
axes[0][0].scatter(decomposited[np.where(clusters==4)[0], 0], decomposited[np.where(clusters==4)[0], 1], 
                  c='b', label='4', alpha=0.3)
axes[0][0].scatter(decomposited[np.where(clusters==5)[0], 0], decomposited[np.where(clusters==5)[0], 1], 
                  c='orange', label='5', alpha=0.3)

axes[0][0].set_title('PCA | EVR: {:.3f}'.format(pca.explained_variance_ratio_.sum()))
axes[0][0].legend()

decomposited = decomposition.KernelPCA(n_components=2, random_state=42).fit_transform(X1.values)
axes[0][1].scatter(decomposited[:,0], decomposited[:,1], c=clusters, cmap=cmap.prism)
axes[0][1].set_title('KernelPCA')

decomposited = decomposition.SparsePCA(n_components=2, random_state=42).fit_transform(X1.values)
axes[1][0].scatter(decomposited[:,0], decomposited[:,1], c=clusters, cmap=cmap.prism)
axes[1][0].set_title('SparsePCA')

svd = decomposition.TruncatedSVD(n_components=2, random_state=42)
decomposited = svd.fit_transform(X1.values)
axes[1][1].scatter(decomposited[:,0], decomposited[:,1], c=clusters, cmap=cmap.jet)
axes[1][1].set_title('TruncatedSVD | EVR: {:.3f}'.format(svd.explained_variance_ratio_.sum()))
```
